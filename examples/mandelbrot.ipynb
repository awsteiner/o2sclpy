# A fun Mandelbrot playground using O$_2$sclpy

```python
import o2sclpy
import matplotlib.pyplot as plot
import sys

plots=True
if 'pytest' in sys.modules:
    plots=False
```

Link the o2scl library:

```python
link=o2sclpy.linker()
link.link_o2scl()
```

```python
f=o2sclpy.fract(link)
```

```python
ug=o2sclpy.uniform_grid_end.init(link,-2,2,199)
t3d=o2sclpy.table3d(link)
rx=o2sclpy.std_vector(link)
ry=o2sclpy.std_vector(link)
rx.resize(4)
ry.resize(4)
rx[0]=-1
rx[1]=1
rx[2]=0
rx[3]=0
ry[0]=0
ry[1]=0
ry[2]=-1
ry[3]=1
vmin=o2sclpy.std_vector(link)
vmax=o2sclpy.std_vector(link)
f.nrf_z4m1(ug,ug,1000,1.0e9,t3d,rx,ry,vmin,vmax)
```

```python
pl=o2sclpy.plotter()
pl.den_plot_temp(t3d,'root')
```

```python
import ctypes
ugx=o2sclpy.uniform_grid_end.init(link,-1.7,0.5,599)
ugy=o2sclpy.uniform_grid_end.init(link,-1.1,1.1,599)
t3d=o2sclpy.table3d(link)
min2=0
max2=0
f.itf_mandel(ugx,ugy,100,10.0,t3d,min2,max2)
```

```python
pl=o2sclpy.plotter()
pl.colbar=1
pl.den_plot_temp(t3d,'it')
```
